```python
import skimage
import numpy as np
from matplotlib import pyplot as plt
from skimage import io
%matplotlib inline
```

```python
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray

#The Otsu method is a simple heuristic to find a threshold to separate the foreground from the background.
def dosegmentationusingotsu(data):
    
    # convert rgb image to gray
    data_gray=rgb2gray(data)
    
    #find the threshold and apply
    threshold = threshold_otsu(data_gray)
    filtered_data = data_gray > threshold
    
    fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
    ax = axes.ravel()
    ax[0].imshow(data_gray, cmap=plt.cm.gray)
    ax[0].set_title('Original image')
    ax[0].axis('off')
    ax[1].imshow(filtered_data, cmap=plt.cm.gray)
    ax[1].set_title('Segmentated result using otsu method')
    ax[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
from skimage.filters import threshold_mean
from skimage.color import rgb2gray

def dosegmentationusingthresholdmean(data):
    # convert rgb image to gray
    data_gray=rgb2gray(data)
    
    #find the threshold and apply
    threshold = threshold_mean(data_gray)
    filtered_data = data_gray > threshold
    
    fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
    ax = axes.ravel()
    ax[0].imshow(data_gray, cmap=plt.cm.gray)
    ax[0].set_title('Original image')
    ax[0].axis('off')
    ax[1].imshow(filtered_data, cmap=plt.cm.gray)
    ax[1].set_title('Segmentated result using mean method')
    ax[1].axis('off')
    
    plt.tight_layout()
    plt.show()
     
```

```python
from skimage.filters import threshold_minimum
from skimage.color import rgb2gray

def dosegmentationusingthresholdminimum(data):
    # convert rgb image to gray
    data_gray=rgb2gray(data)
    
    #threashold the threshold and apply
    threashold = threshold_minimum(data_gray)
    filtered_data = data_gray > threashold
    
    fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
    ax = axes.ravel()
    ax[0].imshow(data_gray, cmap=plt.cm.gray)
    ax[0].set_title('Original image')
    ax[0].axis('off')
    ax[1].imshow(filtered_data, cmap=plt.cm.gray)
    ax[1].set_title('Segmentated Result using minimum method')
    ax[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
from skimage.filters import threshold_local
from skimage.color import rgb2gray

def dosegmentationusingthresholdlocal(data):
    # convert rgb image to gray
    data_gray=rgb2gray(data)
    block_size = 15
    
    #threashold the threshold and apply
    local_threshold = threshold_local(data_gray, block_size, offset=8)
    filtered_data = data_gray > local_threshold
    
    fig, axes = plt.subplots(ncols=2, figsize=(8, 3))
    ax = axes.ravel()
    ax[0].imshow(data_gray, cmap=plt.cm.gray)
    ax[0].set_title('Original image')
    ax[0].axis('off')
    ax[1].imshow(filtered_data, cmap=plt.cm.gray)
    ax[1].set_title('Segmentated Result using local method')
    ax[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# read the images
data = io.imread('mostly_pennies.png')

# call implementation of segmentation using ostu method
dosegmentationusingotsu(data)
# call implementation of segmentation using mean method 
dosegmentationusingthresholdmean(data) 
# call implementation of segmentation using minimum method
dosegmentationusingthresholdminimum(data) 
# call implementation of segmentation using local method
dosegmentationusingthresholdlocal(data)
```
